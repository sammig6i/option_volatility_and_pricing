1. When a stock option is exercised, regardless of the current stock price, the underlying stock is bought or sold at a price equal to the exercise price.

  If the underlying contract is 100 shares of stock, what is the stock position (+ for long, – for short) and cash flow (+ for a credit, – for a debit) resulting from each of the following actions?



Screenshot 2024-05-09 at 7.14.40 PM.png

```python
# a 
stock_position = 200
cash_flow = 70 * 200

print(f'a: Stock position = +{stock_position}')
print(f'a: Cash flow = -{cash_flow}')
print("-" * 10)


# b
stock_position = 800
cash_flow = 40 * 800

print(f'b: Stock position = -{stock_position}')
print(f'b: Cash flow = +{cash_flow}')
print("-" * 10)


# c
stock_position = 16*100
cash_flow = 55 * stock_position

print(f'c: Stock position = -{stock_position}')
print(f'c: Cash flow = +{cash_flow}')
print("-" * 10)


# d
stock_position = 5*100
cash_flow = 120 * stock_position

print(f'd: Stock position = +{stock_position}')
print(f'd: Cash flow = -{cash_flow}')
print("-" * 10)


# e
stock_position = 37*100
cash_flow = 25 * stock_position

print(f'e: Stock position = -{stock_position}')
print(f'e: Cash flow = +{cash_flow}')
print("-" * 10)


# f
stock_position = 21*100
cash_flow = 160 * stock_position

print(f'f: Stock position = -{stock_position}')
print(f'f: Cash flow = +{cash_flow}')
print("-" * 10)


```
